# here u will find neural network implementation from scratch :)

```python
def relu(z):
    return z*(z>0)
```

```python
def sigmoid(z):
    epsilon = np.finfo(z.dtype).eps
    return 1 / (1 + np.exp(-z + epsilon))
```

```python
class neurons:
    def __init__(self,shape,activation):
        self.activation=activation
        self.w=np.random.randn(shape[1],shape[0])*np.sqrt(2./shape[0])
        self.b=np.zeros((shape[1],1))
        #v is for veocity for momentum
        self.v=dict()
        #s is for RMSprop
        self.s=dict()
        self.v_correct=dict()
        self.s_correct=dict()
        self.v["w"]=np.zeros((shape[1],shape[0]))
        self.v["b"]=np.zeros((shape[1],1))
        self.s["w"]=np.zeros((shape[1],shape[0]))
        self.s["b"]=np.zeros((shape[1],1))
        # v_correct because we need to correct the bias in ADAM
        self.v_correct["w"]=np.zeros((shape[1],shape[0]))
        self.v_correct["b"]=np.zeros((shape[1],1))
        self.v_correct["w"]=np.zeros((shape[1],shape[0]))
        self.v_correct["b"]=np.zeros((shape[1],1))
        #gradient is for backpropagation 
        self.gradient=dict()
        self.param=dict()
    def forward(self,input_):
        # Z=W*X +B
        z=np.dot(self.w,input_) + self.b
        if self.activation=="relu":
            # a =z if (z > 0) else 0
            a=relu(z)
        elif self.activation=="sigmoid":
            #a=1/(1+(e^-z))
            a=sigmoid(z)
        else:
            a=z
        #save the data for backpropagation
        self.param["z"]=z
        self.param["a"]=a
        self.param["input"]=input_
        return self.param , a
    def backward(self,da,m):
        # note it's backpropagation for 1 layer only !!!!!!!!
        if self.activation=="relu":
            dz=da* (self.param["z"]>0)
        elif self.activation=="sigmoid":
            dz=da*(self.param["a"]* (1-self.param["a"]) )
        else:
            #linear activatoin
            dz=da
        self.gradient["dw"]=np.dot(dz,self.param["input"].T)/m
        self.gradient["da_prev"]=np.dot(self.w.T,dz)
        self.gradient["db"]=np.array(np.sum(dz,axis=1)/m).reshape(-1,1)
        
        return self.gradient["da_prev"]
    def update(self,lr,B1,B2,t):
        eps=1e-10
        self.v["w"]=(B1*self.v["w"])+((1-B1)*self.gradient["dw"])
        self.v_correct["w"]=self.v["w"]/(1-(B1**t))
        self.v["b"]=(B1*self.v["b"])+((1-B1)*self.gradient["db"])
        self.v_correct["b"]=self.v["b"]/(1-(B1**t))
        self.s["w"]=(B2*self.s["w"])+((1-B2)*(self.gradient["dw"]**2))
        self.s_correct["w"]=self.s["w"]/(1-(B2**t))
        self.s["b"]=(B2*self.s["b"])+((1-B2)*(self.gradient["db"]**2))
        self.s_correct["b"]=self.s["b"]/(1-(B2**t))
        self.w=self.w-lr*(self.v_correct["w"]/(np.sqrt(self.s_correct["w"])+eps))
        self.b=self.b-lr*(self.v_correct["b"]/(np.sqrt(self.s_correct["b"])+eps))
        # u can use normal gradient descent by this 2 comments
#         self.w=self.w-lr*self.gradient["dw"]
#         self.b=self.b-lr*self.gradient["db"]
```

```python
class neural:
    def __init__(self,arch,lr,lamda=0,B1=0.6,B2=0.66,batch_size=16):
        self.arch=arch # neural network architecture (num of layers , neurons ,.....etc)
        self.lr=lr # learning rate (a)
        #self.lamda=lamda
        self.B1=B1
        self.B2=B2
        self.batch_size=batch_size
    def make_batchs(self,x,y):
        #
        # this function create mini batches to use them during training phase
        #
        batch_size=self.batch_size
        mini_batches=[]
        permutation = list(np.random.permutation(x.shape[1]))
        x_shuffle = x[:, permutation]
        y_shuffle = y[:, permutation]
        num_complete_minibatches = int(np.floor(x.shape[1] / batch_size))
        for k in range(0, int(num_complete_minibatches)):
            mini_batch_X = x_shuffle[ : , k * batch_size: (k+1)*(batch_size) ]
            mini_batch_Y = y_shuffle[ : , k * batch_size: (k+1)*(batch_size) ]
            mini_batch = (mini_batch_X, mini_batch_Y)
            mini_batches.append(mini_batch)
        if x.shape[1] % batch_size != 0:
            mini_batch_X = x_shuffle[:, num_complete_minibatches * batch_size:]
            mini_batch_Y = y_shuffle[:, num_complete_minibatches * batch_size:]
        mini_batch = (mini_batch_X, mini_batch_Y)
        mini_batches.append(mini_batch)
        self.data=mini_batches
    def print_nn(self,print_w):
        for node in self.arch:
            print(f"{node.w.shape} <-> {node.activation}")
            if print_w==1:
                print(f"{node.w}")
            print("-----------------------------------------------------")  
    def loss(self,A,y):
        # define Binary cross-entropy loss
        a=A.T
        eps = 1e-10 # epsilon to avoid infinite results
        loss = (-(np.dot(y,np.log(a + eps)) + np.dot((1-y),np.log(1-a + eps))))/self.batch_size
        return loss
    def d_loss(self,a,y):
        # define Binary cross-entropy loss derivative
        eps = 1e-10 # epsilon to avoid infinite results
        return (-y/(a+eps))+((1-y)/(1-a+eps))
    def forward(self,batch):
        #loop over neural network architecture for forward propagation (start from input layer)
        an,a=self.arch[0].forward(batch)
        for i in range(1,len(self.arch)):
            an,a=self.arch[i].forward(a)
        return a
    def backward(self,a,y):
        da=self.d_loss(a,y)
        da=self.arch[-1].backward(da,self.batch_size)
        #loop over neural network architecture for backward propagation in reversed way (start from output layer)
        for i in reversed(range(0,len(self.arch)-1)):
            da= self.arch[i].backward(da,self.batch_size)
    def fit(self,x_train_fun,y_train_fun,validation,epoch=1):
        self.make_batchs(x_train_fun.T,y_train_fun.T)
        for i in range(epoch+1):
            t=0 # this variable is for bais correction
            all_examples=np.array([[ ]])
            y_all=np.array([[ ]])
            for x,y in self.data:
                a=self.forward(x)
                all_examples=np.append(all_examples,a)
                y_all=np.append(y_all,y)
                self.backward(a,y)
                t+=1
                for j in range(0,len(self.arch)):
                    self.arch[j].update(self.lr,self.B1,self.B2,t)
            all_examples_binary = np.where(all_examples >= 0.5, 1, 0)
            if (i%(epoch/10))==0:
                print(f"epoch : {(i)} --- loss = {self.loss(all_examples_binary,y_all)}")
                acc=accuracy_score(y_all.T,all_examples_binary.T.astype("int"))
                print(f"accuracy : {acc}")
                predictions=self.predict(validation[0])
                print(f"val accuracy : {accuracy_score(validation[1],predictions)}")
                print("------------------------------------------------")
        self.score=acc
    def predict(self,batch):
        a=self.forward(batch.T)
        a=np.where(a >= 0.5, 1, 0)
        return a.T
```

```python
#make the shape of input as follow
# for neuron one must be N X M which N is number of feature || M is number of neurons
# for //     two // be N2 X M2 which N2 is number of neurons in previous layer || M2 is number of neurons in current layer
# for //     two // be Nn X Mn which Nn is number of neurons in previous layer || Mn is number of neurons in current layer
model=neural(
    [
        # here u can define neural network architecture 
        # by create objects from neurons class 
        # and pass it as layers with it's parameters shape to initialize them
        neurons ([x.shape[1],20],"relu"),
        neurons ([20,15],"relu"),
        neurons ([15,10],"sigmoid"),
        neurons ([10,1],"sigmoid"),
        
    ],lr=0.00001,batch_size=2,B1=0.5,B2=0.55)
# if u know adam optimization then u should know B1,B2 what refers to :)
```
